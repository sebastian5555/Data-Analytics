### Accounts with IB price

```python
IB_PRICE = pd.read_csv(r'C:\Users\carloma5\Mockup\ib_coverage_df.csv', header=0)
# Using only CR
IB_PRICE = IB_PRICE[IB_PRICE['ID TYPE'] == 'CR']
IB_PRICE = pd.merge(df2,IB_PRICE[['ID','IB Value','IB Covered (%)']], how='left', left_on= 'PARTY_ID', right_on= 'ID')
# Erase rows when IB Value is nan
IB_PRICE = IB_PRICE[~IB_PRICE['IB Value'].isna()]
IB_PRICE
```

```python
df2['RESOLUTION_CODE'].value_counts()
```

```python
# Checks in which quantile the column goes and filter df by only that quartile
M_IB_PRICE = IB_PRICE['IB Value'][IB_PRICE['PARTY_ID'] == 38913].min()
def check_quantile(df):
    if df['IB Value'].quantile(0.25) > M_IB_PRICE: 
        df = df[df['IB Value'] >= df['IB Value'].quantile(0.25)]
    elif df['IB Value'].quantile(0.50) > M_IB_PRICE: 
        df = df[df['IB Value'] >= df['IB Value'].quantile(0.50)]
    else:
        df = df[df['IB Value'] >= df['IB Value'].quantile(0.75)]
    return df
```

### Volumetry

```python
general_cases = df2.shape[0]
microsoft_cases = df.shape[0]
v_df = check_quantile(IB_PRICE)
v_df = v_df.groupby(['PARTY_ID']).size().reset_index(name='count')
print('Industry cases: {0}'.format(v_df['count'].sum()/v_df.shape[0]), '\nMicrosoft cases: {0}'.format(microsoft_cases))
```

```python
fig, ax = plt.subplots()
sns.boxplot(v_df, y='count',showfliers=True,
            width=0.3, boxprops={'zorder': 1, 'alpha': 0.4}, fliersize=3,
            flierprops=dict(markeredgecolor='moccasin'),
            medianprops=dict(color='r'))
ax.plot(microsoft_cases, 'X', markersize= 8, color='indigo', label='Microsoft')
#plt.legend()
plt.show()
```
